```python
import numpy as np                         # Sayısal işlemler için
import pandas as pd                        # Veri okuma ve düzenleme
import seaborn as sns                      # Veri görselleştirme
import matplotlib.pyplot as plt            # Grafik çizimi
from matplotlib.cbook import boxplot_stats # Aykırı değer tespiti için
import re                                  # Veri temizliinde kolaylık için

from scipy.stats import pearsonr           # Korelasyon analizi
from sklearn.metrics import mean_squared_error, mean_absolute_error  # Hata metrikleri

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

print("✅ Gerekli kütüphaneler başarıyla yüklendi.")
```

veriseti düzenlenmesi;


```python
from google.colab import drive
drive.mount('/content/drive')

dosya_yolu = '/content/drive/MyDrive/Colab Notebooks/SalaryPrediction.csv'

data = pd.read_csv(dosya_yolu)


#Verisetindeki boşlukların temizlenmesi

data.isnull().sum()


data.dropna(inplace=True)

#doğrudan maaş bilgisini taşıdığı için Job Title değişkeni çıkarılmıştır.
#Bu sayede modelin veri sızıntısından etkilenmesi önlenmiş ve genel performansı artırılmıştır.

data = data.drop("Job Title", axis=1)

#kategorik verilerin sayısallaştırılması

data_encoded = pd.get_dummies(data, columns=["Gender", "Education Level"], drop_first=True)


#Verilerin hedef ve özellikler olarak ayrılması

X = data_encoded.drop("Salary", axis=1) #X, Gelir harici sütunları alır
y = data_encoded["Salary"] #Y, Gelir sütununu alır

#Verilerin Test ve Eğitim olarak ayrılması

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

modelin kurulumu ve eğitimi;


```python
model = LinearRegression()
model.fit(X_train, y_train)
```

modelin tahmini;

```python
y_pred = model.predict(X_test)
```

R² değeri ve Kök Ortalama Kare Hatası;


```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("R²:", r2)
```

Her değişkenin Yılık Gelire etkisi;


```python
coefficients = pd.DataFrame(model.coef_, X.columns, columns=["Coefficient"])
print(coefficients)
```
